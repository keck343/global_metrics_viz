```python
health_unviersal = pd.read_csv("healthcae_and_metrics_nona.csv")
health_unviersal = health_unviersal.drop(columns=['Unnamed: 0'])
health_unviersal.head()
```

```python
def unversial(row):
    if row != 'None':
        return "Universal Coverage"
    else:
        return "No Universal Coverage"

health_unviersal['Healthcare'] = health_unviersal.Type.apply(lambda row: unversial(row))

health_unviersal.head()
```

```python
health_unviersal.columns = ['Name', 'Infant_Mortality', 'Credit_Rating', 'GDP', 'Women_in_Labor_Force', 
                            'Year', 'Type', 'Heathcare']
```

```python
health_unviersal.to_csv("health_unviersal_rename_cols.csv")
```
